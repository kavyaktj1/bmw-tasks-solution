1. Write a function that takes as input two timestamps of the form 2017/05/13 12:00 and
calculates their differences in hours. Please only return the full hour difference and
round the results. E.g. 2022/02/15 00:05 and 2022/02/15 01:00 would return 1 hour.

2. Expand the above function to only count the time difference between 09:00 – 17:00 and
only on weekdays.

```python
from datetime import datetime

#Takes two timestamp strings in the format 'YYYY/MM/DD HH:MM', Converts them to datetime objects
#Calculates the absolute difference in hours, Rounds to the nearest whole hour, Returns the result as an integer

def hour_difference(timestamp1, timestamp2):
    # Convert string timestamps to datetime objects
    t1 = datetime.strptime(timestamp1, '%Y/%m/%d %H:%M')
    t2 = datetime.strptime(timestamp2, '%Y/%m/%d %H:%M')

    # Calculate time difference in hours
    diff = abs(t2 - t1).total_seconds() / 3600

    # Round to nearest full hour
    return round(diff)
```

```python
# Example 1
print(hour_difference('2022/02/15 00:05', '2022/02/15 01:00'))  # Output: 1

# Example 2
print(hour_difference('2017/05/13 12:00', '2017/05/14 15:30'))  # Output: 4
```

Function to only count the time difference between 09:00–17:00 on weekdays (Monday to Friday). The function will exclude weekends and hours outside the 9 AM to 5 PM window, calculating the effective time difference in hours and rounding to the nearest whole hour.

```python
from datetime import datetime, time, timedelta

def hour_difference(timestamp1, timestamp2):
    # Convert string timestamps to datetime objects
    t1 = datetime.strptime(timestamp1, '%Y/%m/%d %H:%M')
    t2 = datetime.strptime(timestamp2, '%Y/%m/%d %H:%M')

    # Ensure t1 is earlier than t2 for simplicity
    if t1 > t2:
        t1, t2 = t2, t1

    # Define business hours (09:00 to 17:00)
    start_time = time(9, 0)
    end_time = time(17, 0)
    total_hours = 0

    # Initialize current time to t1
    current = t1

    #Weekday Check: Uses weekday() (0=Monday, 6=Sunday) to skip weekends (Saturday and Sunday).
    while current < t2:
        # Check if current day is a weekday (Monday=0, Sunday=6)
        if current.weekday() >= 5:  # Skip Saturday and Sunday
            current = current.replace(hour=0, minute=0, second=0, microsecond=0) + timedelta(days=1)
            continue

        # Define start and end of business hours for the current day
        day_start = datetime.combine(current.date(), start_time)
        day_end = datetime.combine(current.date(), end_time)

        # Adjust start and end based on actual timestamps
        period_start = max(current, day_start)
        period_end = min(t2, day_end)

        # Only count time if period_start is within business hours and before period_end
        if period_start < period_end and period_start <= day_end:
            hours = (period_end - period_start).total_seconds() / 3600
            total_hours += hours

        # Move to the next day
        current = current.replace(hour=0, minute=0, second=0, microsecond=0) + timedelta(days=1)

    # Round to nearest full hour
    return round(total_hours)
```
